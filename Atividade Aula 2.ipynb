```python
pessoas.head(10)
```

```python
n = pessoas.shape[0]
```

```python
df1 = pd.DataFrame()
df1['Regiao']= pessoas.UF
df1.Regiao[(pessoas.UF>=10) & (pessoas.UF<20)]=1
df1.Regiao[(pessoas.UF>=20) & (pessoas.UF<30)]=2
df1.Regiao[(pessoas.UF>=30) & (pessoas.UF<40)]=3
df1.Regiao[(pessoas.UF>=40) & (pessoas.UF<50)]=4
df1.Regiao[(pessoas.UF>=50) & (pessoas.UF<60)]=5

df1['Regiao'] = df1['Regiao'].astype('category')
df1.Regiao.cat.categories = ('Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro_Oeste')
```

```python
pessoas['V6002'] = pessoas['V6002'].astype('category')
pessoas.V6002.cat.categories = ('Publico', 'Privado', 'NA')

pessoas.V4707.cat.categories = ('Até 14 horas','15 a 39 horas','40 a 44 horas','45 a 48 horas','49 horas ou mais','NA')
pessoas.V4707 = pessoas.V4707.astype('category') 

pessoas['V4802'] = pessoas['V4802'].astype('category')
pessoas.V4802.cat.categories = ('Maternal','Alfabetização', 'Alfabetização dos adultos', '1ª a 4ª séries',
                                '5ª a 8ª séries','1ª série','2ª a 5ª séries','6ª a 9ª séries',
                                'não seriado do ensino fundamental', 'Ensino médio','Superior', 'Não Aplicavel')

pessoas['UF'] = pessoas['UF'].astype('category')
pessoas.UF.cat.categories = ("Rondonia", "Acre", "Amazonas", "Roraima", "Pará","Amapá",
                                "Tocantins", "Maranhão", "Piauí", "Ceará", "Rio Grande do Norte",
                                "Paraíba", "Pernambuco", "Alagoas", "Sergipe", "Bahia", "Minas Gerais",
                                "Espírito Santo", "Rio de Janeiro", "São Paulo", "Paraná", "Santa Catarina",
                                "Rio Grande do Sul", "Mato Grosso do Sul", "Mato Grosso", "Goiás", "Distrito Federal")

pessoas['V0404'] = pessoas['V0404'].astype('category')
pessoas['V0302'] = pessoas['V0302'].astype('category')
pessoas.V0404.cat.categories = ('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda', 'Sem Dec')
pessoas.V0302.cat.categories = ('Masculino', 'Feminino')
```

```python
print('Frequências relativas por Sexo:')
ut1 = (pessoas.V0302.value_counts()/n*100)
print(ut1)
```

```python
plot = ut1.plot(kind='pie',title='Raça geral da população',autopct='%.1f',figsize=(6, 6))
```

```python
print('Frequências relativas por Raça:')
ut2 = (pessoas.V0404.value_counts()/n*100)
print(ut2)
```

```python
plot = ut2.plot(kind='pie',title='Raça geral da população',autopct='%.1f',figsize=(6, 6))
```

```python
plot = ut2.plot(kind='bar',title='População por raça',figsize=(6, 6),color=('b','g','r','c','m','y'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Satisfação')
```

```python
print('Frequências relativas por Para cada Sexo:')
ut2FM = (pd.crosstab(pessoas.V0404, pessoas.V0302)/n*100)
print(ut2FM)
```

```python
plot = ut2FM.plot(kind='pie',title='Tipo de educação',autopct='%.1f',figsize=(13, 6), subplots=True)
```

```python
print('Frequências relativas por Instituição:')
ut3= (pessoas.V6002.value_counts()/n*100)
print(ut3)
```
